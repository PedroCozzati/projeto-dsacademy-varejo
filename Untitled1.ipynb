```python
dataset['Mes'] = dataset['Data_Pedido'].dt.month
```

```python
question9 = dataset.groupby(['Ano','Mes',"Segmento"])['Valor_Venda'].agg([np.sum,np.mean,np.median])
```

```python
question9
```

```python
anos = question9.index.get_level_values(0)
meses = question9.index.get_level_values(1)
segmentos = question9.index.get_level_values(2)
```

```python
plt.figure(figsize=(12,6))
sea.set()
fig1 = sea.relplot(kind='line',data=question9,y='mean',x=meses,hue=segmentos,col=anos,col_wrap=2)
fig1.set_titles('MÃ©dia de vendas por segmento por ano e por mÃªs')
```
